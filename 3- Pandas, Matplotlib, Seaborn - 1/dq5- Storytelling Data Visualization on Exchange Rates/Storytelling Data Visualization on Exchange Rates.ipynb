## Data Cleaning

#### Rename the cols US dollar and Period\Unit to US_dollar and Time.


```python
exchange_rates.rename(columns={'[US dollar ]': 'US_dollar',
                               'Period\\Unit:': 'Time'},
                      inplace=True)
```

```python
exchange_rates['Time'] = pd.to_datetime(exchange_rates['Time'])
```

```python
exchange_rates.sort_values('Time', inplace=True)
exchange_rates.reset_index(drop=True, inplace=True)
```

#### Isolating the Time and US_dollar columns to a separate DataFrame

```python
euro_to_dollar = exchange_rates[['Time', 'US_dollar']].copy()
```

```python
# Lets check if it ran nicely
euro_to_dollar['US_dollar'].value_counts()
```

**Observation**:
- As we can see '-' character appearing in our Series, lets consider dropping any rows that contain it 

```python
euro_to_dollar = euro_to_dollar[euro_to_dollar['US_dollar'] != '-']
```

#### Converting the US_dollar column to a float data type

```python
euro_to_dollar['US_dollar'] = euro_to_dollar['US_dollar'].astype(float)
```

## Data Visualization

```python
plt.plot(euro_to_dollar['Time'],
         euro_to_dollar['US_dollar'])
plt.show()
```

#### Lets consider using 'Rolling Mean' a.k.a 'Moving Averages' for the visualization

```python
euro_to_dollar['rolling_mean'] = euro_to_dollar['US_dollar'].rolling(30).mean()
```

```python
euro_to_dollar
```

```python
plt.figure(figsize=(12,9))

plt.subplot(3,2,1)
plt.plot(euro_to_dollar['Time'], euro_to_dollar['US_dollar'])
plt.title('Original values', weight='bold')

for days, rolling_mean in zip([2, 3, 4, 5, 6],
                           [7, 30, 90, 180, 360]):
    plt.subplot(3,2, days)
    plt.plot(euro_to_dollar['Time'],
             euro_to_dollar['US_dollar'].rolling(rolling_mean).mean())
    plt.title('Rolling Window: ' + str(rolling_mean) + ' (days)')
    
plt.tight_layout() 
plt.show()
```

#### Lets see how the euro-dollar rate has changed during the coronavirus pandemic. We can show the 2020 data and the 2016-2019 data as a baseline using a line plot.

<font color='green'> Thanks **@adetolaadeya** for your wonderful piece. It was super helpful for me to to create the Visualization below

```python
during_corona = euro_to_dollar.copy()[euro_to_dollar['Time'].dt.year == 2020]
before_corona = euro_to_dollar.copy()[(euro_to_dollar['Time'].dt.year >= 2016) & (euro_to_dollar['Time'].dt.year <= 2019)]
```

```python
# Choose Style 
style.use('seaborn-darkgrid')

# Add subplots
plt.figure(figsize=(11,6))
ax1 = plt.subplot(2,2,1)
ax2 = plt.subplot(2,2,2)
ax3 = plt.subplot(2,1,2)

axes = [ax1, ax2, ax3]


# Customizing the subplots
for ax in axes:
    ax.set_ylim(0.8, 1.3)
    ax.set_yticks([0.9, 1.0, 1.1, 1.2])
    ax.set_yticklabels(['0.9', '1.0', '1.1', '1.2'],
                   alpha=0.4)
    ax.grid(alpha=0.7) 
    
# ax1 Pre Covid Situation, from 2016 to 2019
ax1.plot(before_corona['Time'], before_corona['rolling_mean'],
         color='#15B01A')
ax1.set_xticklabels(['', '2016', '', '2017', '', '2018', '',
                     '2019'],
                   alpha=0.5)
#ax1.get_xticks()
#ax1.get_yticks()

ax1.text(736250, 1.35,'Pre-COVID19 (2016-2019)', fontsize=12, weight='bold',
        color='#15B01A')

# ax2: During the Covid Situation, year 2020
ax2.plot(during_corona['Time'], during_corona['rolling_mean'],
        color='#E50000')
ax2.set_xticklabels(['Q1-2020', '', 'Q2-2020', '', 'Q3-2020', '',
                     'Q4-2020'],
                   alpha=0.5)
#ax2.get_xticks()
#ax2.get_yticks()
ax2.text(737500, 1.35, 'During COVID19 (2020)', fontsize=12, weight='bold',
        color='#E50000')


# ax3: Complete picture
ax3.plot(before_corona['Time'], before_corona['rolling_mean'],
        color='#15B01A')
ax3.plot(during_corona['Time'], during_corona['rolling_mean'],
        color='#E50000')
ax3.grid(alpha=0.5)
ax3.set_xticks([]) 

### Headlines
ax1.text(735700, 1.55, 'EURO-USD rates peaked at 1.21 during the Pandemic',
         fontsize=22, weight='bold')
ax1.text(735700, 1.45, '''Before the Covid-19 era, the rate topped at 1.24'''
,fontsize=16)
ax1.get_xticks()
ax1.get_yticks()
```

## Conclusions:

- We can see the EUR-USD rate tended to decline gradually through Q1 to Q2.
- Q2 seen a steady but significant increament in the rate and the rise continued till the beginning of Q3
- There was a slight decrease in the rate during Q3
- In Q4, the rate recovered from the decline it faced in Q3 and by the end of Q4, the rate managed to touch its peak at 1.21
